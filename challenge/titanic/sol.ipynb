```python
df['cabin_mapped'] = df['Cabin'].map(lambda x: str(x)[0])
```

```python
cat_columns = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked', 'cabin_mapped']

for i in cat_columns:
    print(i)
    grouped = df.groupby(i).aggregate({"Survived": "sum"})
    grouped = pd.concat([grouped, df[i].value_counts()], axis=1)
    grouped.rename(columns={i: "Total"}, inplace=True)
    grouped['Percentage'] = grouped['Survived'] / grouped['Total']
    display(grouped)
```
